```python
# GPT 1-1ë²ˆ
subjects = ["êµ­ì–´", "ì˜ì–´", "ìˆ˜í•™", "ì‚¬íšŒ", "ê³¼í•™"]

# ì‰¼í‘œë¡œ êµ¬ë¶„ëœ ì ìˆ˜ ì…ë ¥
scores_str = input("ì ìˆ˜ 5ê°œë¥¼ ì‰¼í‘œë¡œ êµ¬ë¶„í•´ ì…ë ¥í•˜ì„¸ìš”: ")  # ì˜ˆ: 76,88,62,94,78

# ë¬¸ìì—´ â†’ ì •ìˆ˜ ë¦¬ìŠ¤íŠ¸ ë³€í™˜
scores = [int(s.strip()) for s in scores_str.split(",")]

# ìµœì € / ìµœê³  ì°¾ê¸°
min_score = min(scores)
max_score = max(scores)

min_subject = subjects[scores.index(min_score)]
max_subject = subjects[scores.index(max_score)]

# í‰ê· 
avg = sum(scores) / len(scores)

# ê²°ê³¼ ì¶œë ¥
print(f"ìµœì € ì ìˆ˜: {min_score} ({min_subject})")
print(f"ìµœê³  ì ìˆ˜: {max_score} ({max_subject})")
print(f"í‰ê·  ì ìˆ˜: {avg:.2f}")
```
