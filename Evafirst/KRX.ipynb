```python
import matplotlib.pyplot as plt # 산점도를 시각화 해주는 라이브러리
col1 = 'date'
col2 = 'Open'

plt.scatter(df[col1], df[col2])
```

```python
df = df.sort_values(by='date', ascending=True)
plt.scatter(df[col1], df[col2])

plt.xticks([])
plt.yticks([])
plt.xlabel('date')
plt.ylabel('Open')
```

```python
plt.scatter(df['date'], df['Close'])

plt.xticks([])
plt.yticks([])
plt.xlabel('date')
plt.ylabel('Close')
```

```python
plt.scatter(df['date'], df['Open'])

plt.xlabel('date')
plt.ylabel('Open')
# 원하는 눈금 개수 설정
num_ticks = 10

# len(df['date])은 날짜 열 길이
# 데이터의 전체길이를 눈금 개수로 나눈 값이 각 눈금 사이의 간격
tick_positions = range(0, len(df['date']), len(df['date']) // num_ticks)

# 눈금 위치에 해당하는 날짜 값을 가져와 사용
tick_labels = [df['date'][i] for i in tick_positions]

# x축에 눈금을 설정하고, 45도 회전하여 눈금 레이블이 겹치지 않도록 함.
plt.xticks(tick_positions, tick_labels, rotation=45)

# 그림을 표시한다.
plt.show()
```

```python
df =  fdr.DataReader('005930', '2010-01-01')
df.head(5)
```

```python
df1=df.loc['2022-01-01':'2022-12-31']
df1['ret']=(df1['Close'].pct_change())*100
df2=df1['ret']
r=df2.dropna()
len(r)
```

```python
positive=0
n=len(r)
for i in range(n):
  if r[i]>0:
    positive+=1
positive
```

```python
positive_rate=positive/n
```

```python
positive_rate
```
