## 구매금액 z-score 계산

```python
# cc_num 컬럼의 값마다 데이터 개수 계산
cc_df['cc_num'].value_counts()
```

```python
# cc_num 컬럼의 값마다 amt의 평균, 표준편차를 계산하여 amt_info에 저장
amt_info = cc_df.groupby('cc_num')['amt'].agg(['mean', 'std']).reset_index()
```

```python
# 정보를 피클로 저장
amt_info.to_pickle('./amt_info_pkl')
```

```python
# cc_num 컬럼을 기준으로, cc_df와 amt_info 데이터를 합쳐서 cc_df에 저장 (left merge)
cc_df = cc_df.merge(amt_info, on = 'cc_num', how = 'left')
```

```python
# 결제금액(amt)의 z-score를 계산하여 amt_z 컬럼에 저장합니다.
cc_df['amt_z'] = (cc_df['amt'] - cc_df['mean']) / cc_df['std']
```

```python
# z-score 계산이 완료되었으니 mean, std 컬럼을 제거
cc_df.drop(['mean','std'], axis = 1, inplace = True)
```

```python
# cc_num, category 컬럼의 값마다 결제금액(amt)의 평균, 표준편차를 계산하여 cat_info에 저장
cat_info = cc_df.groupby(['cc_num','category'])['amt'].agg(['mean','std']).reset_index()
```

```python
# 정보를 피클로 저장
cat_info.to_pickle('./cat_info.pkl')
```

```python
# cc_num 컬럼을 기준으로, cc_df와 cat_info 데이터를 합쳐서 cc_df에 저장 (left merge)
cc_df = cc_df.merge(cat_info, on = ['cc_num','category'], how = 'left')
```

```python
#  결제금액(amt)의 z-score를 계산하여 cat_amt_z 컬럼에 저장
cc_df['cat_amt_z'] = (cc_df['amt'] - cc_df['mean']) / cc_df['std']
```

```python
# z-score 계산이 완료되었으니 mean, std 컬럼을 제거
cc_df.drop(['mean','std'], axis =1 , inplace = True)
```

## 결제 시간 관련 feature 분석

```python
# trans_date_trans_time 컬럼에서 시간 값을 추출하여 hour 컬럼에 저장합니다. (datetime 자료형)
cc_df['hour'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
```

```python
cc_df.tail(10)
```

```python
# 결제시간을 morning, afternoon, night, evening으로 분류하기 위해 함수를 정의
def hour_func(x):
    if (x >= 7) & ( x < 12):
        return 'morning'
    elif (x >= 12) & ( x < 18):
        return 'afternoon'
    elif (x >= 18) & ( x < 22):
        return 'night'
    else:
        return 'evening'
```

```python
# hour 컬럼에 hour_func 함수를 적용한 값을 hour_cat 컬럼에 저장
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
# 시간대별로 데이터 수를 계산하여 출력
cc_df['hour_cat'].value_counts()
```

```python
# cc_num 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 all_cnt 변수로 저장
all_cnt = cc_df.groupby('cc_num')['amt'].count().reset_index()
```

```python
# cc_num, hour_cat 컬럼의 값마다 amt 컬럼의 데이터 수를 계산하여 hour_cnt 변수로 저장
hour_cnt = cc_df.groupby(['cc_num', 'hour_cat'])['amt'].count().reset_index()
```

```python
all_cnt.head()
```

```python
hour_cnt.head()
```

```python
# cc_num 컬럼을 기준으로 hour_cnt와 all_cnt를 합칩니다.
hour_cnt = hour_cnt.merge(all_cnt, on = 'cc_num', how = 'left')
```

```python
hour_cnt.head()
```

```python
# amt_x와 amt_y 컬럼의 이름을 각각 hour_cnt, total_cnt로 변경
hour_cnt = hour_cnt.rename({'amt_x': 'hour_cnt', 'amt_y': 'total_cnt'}, axis = 1)
```

```python
hour_cnt.head()
```

```python
# 전체 거래 건수(total_cnt)에 대한 시간대별 거래 건수(hour_cnt)의 비율을 계산하여 hour_perc 컬럼에 저장
hour_cnt['hour_perc'] = hour_cnt['hour_cnt'] / hour_cnt['total_cnt']
```

```python
hour_cnt.head(10)
```

```python
# 시간대별 모두 합산이 '1'을 확인.
hour_cnt.loc[0:3]['hour_perc'].sum()
```

```python
hour_cnt.tail(10)
```

```python
cc_df.head()
```

```python
# hour_cnt에 cc_num, hour_cat, hour_perc만 남기고 업데이트
hour_cnt = hour_cnt[['cc_num','hour_cat','hour_perc']]
```

```python
# 피클저장
hour_cnt.to_pickle('./hour_cnt.pkl')
```

```python
# cc_num, hour_cat 컬럼을 기준으로 cc_df와 hour_cnt 데이터를 합친다.
cc_df = cc_df.merge(hour_cnt, on = ['cc_num','hour_cat'], how = 'left')
```

```python
cc_df.drop(['trans_date_trans_time', 'hour', 'hour_cat'], axis = 1, inplace = True)
```

## 거리 관련 feature 분석

```python
# geopy 설치
# !pip install geopy 
from geopy.distance import distance
```

```python
# 고객의 위치와 상점의 위치 사이의 거리를 계산하여 distance 컬럼에 저장 
cc_df['distance'] = cc_df.apply(lambda x: distance((x['lat'], x['long']), (x['merch_lat'], x['merch_long'])).km, axis = 1)
```

```python
# cc_num 컬럼의 값마다 distance의 평균, 표준편차를 계산하여 dist_info 변수에 저장
dist_info = cc_df.groupby('cc_num')['distance'].agg(['mean', 'std']).reset_index()
```

```python
# cc_num 컬럼을 기준으로 cc_df와 dist_info 데이터를 합친다.
cc_df = cc_df.merge(dist_info, on = 'cc_num', how = 'left')
```

```python
cc_df.head()
```

```python
# distance 값의 z-score를 계산하여 dist_z 컬럼에 저장
cc_df['dist_z'] = (cc_df['distance'] - cc_df['mean']) / cc_df['std']
```

```python
# 피클 저장
dist_info.to_pickle('./dist_info.pkl')
```
